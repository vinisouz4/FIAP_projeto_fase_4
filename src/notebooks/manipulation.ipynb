## Novas datas para teste do modelo na API

```python
list_symbol = ['AAPL']

start_date = '2026-01-21'
end_date = '2026-05-03'
```

```python
df = yf.download(list_symbol, start=start_date, end=end_date)

df_index_ajusted = df.stack(level=1).reset_index()


df_index_ajusted.columns = [
    "date", "ticker", "open", "high", "low", "close", "volume"
]

df_index_ajusted.sort_values(by=["date"],inplace=True)

df_index_ajusted
```

```python
result = (
    df_index_ajusted
    .groupby("ticker")
    .apply(lambda x: {
        "ticker": x["ticker"].iloc[0],
        "history": x[["open", "high", "low", "close", "volume"]]
        .to_dict(orient="records")
    })
    .tolist()
)

import json

print(json.dumps(result, indent=2))
```
