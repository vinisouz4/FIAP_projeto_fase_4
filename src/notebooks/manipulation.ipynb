```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
import datetime
import pandas as pd
import os
import sys
```

```python
list_symbol = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META']

start_date = '2002-01-01'
end_date = '2026-04-20'
```

```python
df = yf.download(list_symbol, start=start_date, end=end_date)

df_index_ajusted = df.stack(level=1).reset_index()

df_index_ajusted
```
